```python
# -- Plot Template --
# Define general parameters
FONT_FAMILY = 'Arial'
FONT_SIZE = 30
AXIS_WIDTH = 2.5
TICK_WIDTH = 2.5
TICK_LENGTH_MAJOR = 6
TICK_LENGTH_MINOR = 0
FIG_SIZE_DEFAULT = (10, 8)  # size in inches
DPI_DISPLAY = 100
DPI_HIGH_RES = 600

# Apply global setting
font = {'family': FONT_FAMILY, 'size': FONT_SIZE}
matplotlib.rc('font', **font)
matplotlib.rc('axes', linewidth=AXIS_WIDTH)
def set_ticks(ax_list):
    """
    Apply consistent tick settings to a list of matplotlib axes objects.
    
    Args:
        ax_list (list): A list of matplotlib axis objects (e.g., [ax1, ax2]).
    """
    for ax in ax_list:
        ax.tick_params(
            which='major',
            length=TICK_LENGTH_MAJOR,
            width=TICK_WIDTH
        )
        ax.tick_params(
            which='minor',
            length=TICK_LENGTH_MINOR,
            width=TICK_WIDTH
        )
        ax.tick_params(
            which='both',
            axis='both',
            direction='in',
            pad=5,
            right=True,
            top=True
        )
# -- End of Template --
```

```python
def gaussian_envelope_3d(xyz, A_env, x0, y0, z0, sigma_env):
    """
    三维高斯包络函数 (代码波包振幅的空间分布)
    
    Args:
        xyz: (x, y, z) tuple arrays, 原子的平衡位置
        A: 包络函数的幅度 (包络的峰值)
        x0: 包络中心 x 坐标
        y0: 包括中心 y 坐标
        z0: 包络中心 z 坐标
        sigma: 包络函数的宽度 (标准差)
    
    Returns:
        高斯包络函数值数组 (代表空间各点的理论振幅)
    """
    x, y, z = xyz
    envelope = A_env * np.exp(-((x - x0)**2 + (y - y0)**2 + (z - z0)**2) / (2 * sigma_env**2))
    return envelope

def fit_gaussian_envelope_3d(positions, envolope_proxy_values, tolerance=0.01):
    """
    对代表波包包络的物理量分布进行三维高斯包络拟合, 确定包络中心
    
    Args:
        positions (np.arrays): 原子位置数组 (通常是平衡位置 R0), shape: (natoms, 3)
        envolope_proxy_values (np.array): 代表包络强度的物理量值数组 (如瞬时位移大小 |u(t)|), shape: (natoms,)
        tolerance (float): 容差因子, 用于定义范围边界 (default: 0.01)
            
    Returns:
        tuple: (x0_fit, y0_fit, z0_fit, A_env_fit, sigma_env_fit, success)
    """
    x_data = positions[:, 0]
    y_data = positions[:, 1]
    z_data = positions[:, 2]
    f_data = envolope_proxy_values
    
    # 计算坐标范围
    x_min, x_max = np.min(x_data), np.max(x_data)
    y_min, y_max = np.min(y_data), np.max(y_data)
    z_min, z_max = np.min(z_data), np.max(z_data)
    
    # 初始猜测参数
    A_env_guess = np.max(f_data)
    max_idx = np.argmax(f_data)
    x0_guess = x_data[max_idx]
    y0_guess = y_data[max_idx]
    z0_guess = z_data[max_idx]
    sigma_env_guess = np.std(positions, axis=0).mean() / 2 # 使用各维度标准差的平均值
    
    initial_guess = [A_env_guess, x0_guess, y0_guess, z0_guess, sigma_env_guess]
    
    try:
        popt, _ = curve_fit(gaussian_envelope_3d, (x_data, y_data, z_data), f_data, p0=initial_guess, maxfev=100000)  # 可能需要多次迭代
        A_env_fit, x0_fit, y0_fit, z0_fit, sigma_env_fit = popt
        # 检查拟合的中心是否在坐标范围内
        in_range = (x_min - tolerance * (x_max - x_min) <= x0_fit <= x_max + tolerance * (x_max - x_min) and
                    y_min - tolerance * (y_max - y_min) <= y0_fit <= y_max + tolerance * (y_max - y_min) and
                    z_min - tolerance * (z_max - z_min) <= z0_fit <= z_max + tolerance * (z_max - z_min))
        if in_range:
            success = True
            #print(f"  - 3D Gaussian fit succeeded and center is within range (tolerance={tolerance})")        
        else:
            success = False
            #print(f"  - 3D Gaussian fit succeeded but center ({x0_fit:.3f}, {y0_fit:.3f}, {z0_fit:.3f}) outside coordinate range")
    except (RuntimeError, ValueError) as e:
        #print(f"  - Warning: 3D Gaussian fit failed: {e}")
        A_env_fit, x0_fit, y0_fit, z0_fit, sigma_env_fit = np.nan, np.nan, np.nan, np.nan, np.nan
        success = False
    return x0_fit, y0_fit, z0_fit, A_env_fit, sigma_env_fit, success
```

```python
def calculate_energy_transmission(time_indexes, times, time_unit, atom_velocity_magnitudes_per_frame, atom_potential_energies_per_frame,
                                          pristine_atom_potential_energies, masses,
                                          pristine_positions, propagation_direction, interface_position):
    """
    计算指定时间点的能量透射情况
    
    Args:
        time_indexes (list): 时间索引列表
        times (arrays): 每一帧时间
        time_unit (str): 时间单位
        atom_velocity_magnitudes_per_frame (array): 每一帧原子速度大小, shape: (nframes, natoms)
        atom_potential_energies_per_frame (array): 每一帧原子势能, shape: (nframes, natoms)
        pristine_atom_potential_energies: 原始原子势能
        masses (array): 原子质量, shape: (natoms,)
        pristine_positions (array): 原始原子位置, shape: (natoms, 3)
        propagation_direction (str): 波包传播方向 ('x', 'y', 'z' 或 0, 1, 2)
        interface_position (float): 界面位置 (default: None)
    """
    print("\nCalculating energy transmission...")

    # 映射方向字符串到数组索引
    dir_map = {'x': 0, 'y': 1, 'z': 2, 0: 0, 1: 1, 2: 2}
    if propagation_direction not in dir_map:
        raise ValueError(f"Invalid propagation_direction: {propagation_direction}. Use 'x', 'y', 'z' or 0, 1, 2.")
    dir_idx = dir_map[propagation_direction]
    
    
    # 根据指定方向的坐标将原子分为两部分
    coords_in_dir = pristine_positions[:, dir_idx]  # shape: (natoms,)
    if interface_position is not None:
        coord_split = interface_position  # 可直接指定界面位置
    else:
        coord_split = (np.min(coords_in_dir) + np.max(coords_in_dir)) / 2  # 使用中点分割
    left_mask = coords_in_dir <= coord_split
    right_mask = coords_in_dir > coord_split
    
    
    # 计算原始结构的平衡势能
    left_pristine_potential_energy = np.sum(pristine_atom_potential_energies[left_mask])
    right_pristine_potential_energy = np.sum(pristine_atom_potential_energies[right_mask])
    
    
    # 计算每个时间点的波包能量
    for time_idx in time_indexes:
        if time_idx >= len(times):
            continue
        
        # 获取指定时间点的物理量
        t = times[time_idx]
        atom_velocity_magnitudes_t = atom_velocity_magnitudes_per_frame[time_idx]  # shape: (natoms,)
        atom_potential_energies_t = atom_potential_energies_per_frame[time_idx]  # shape: (natoms,)
    
        # 计算左侧部分的波包能量
        left_velocity_magnitudes = atom_velocity_magnitudes_t[left_mask]  # 左侧原子速度大小
        left_masses = masses[left_mask]  # 左侧原子质量
        left_kinetic_energy = 0.5 * np.sum(left_masses * left_velocity_magnitudes**2) * 103.6427  # 左侧动能, unit: eV
    
        left_potential_energy = np.sum(atom_potential_energies_t[left_mask]) - left_pristine_potential_energy # 左侧势能
        left_total_energy = left_kinetic_energy + left_potential_energy  # 左侧总能

        # 计算右侧部分的波包能量
        right_velocity_magnitudes = atom_velocity_magnitudes_t[right_mask]  # 右侧原子速度大小
        right_masses = masses[right_mask]  # 右侧原子质量
        right_kinetic_energy = 0.5 * np.sum(right_masses * right_velocity_magnitudes**2) * 103.6427  # 右侧动能, unit: eV
        
        right_potential_energy = np.sum(atom_potential_energies_t[right_mask]) - right_pristine_potential_energy # 右侧势能, unit: eV
        right_total_energy = right_kinetic_energy + right_potential_energy  # 右侧总能
    
        # 计算能量占比
        if (left_total_energy + right_total_energy) != 0:
            left_ratio = (left_total_energy / (left_total_energy + right_total_energy)) * 100.0
            right_ratio = (right_total_energy / (left_total_energy + right_total_energy)) * 100.0
        else:
            left_ratio = 0.0
            right_ratio = 0.0

        print(f"  - At t = {t} {time_unit}, total energy: {(left_total_energy + right_total_energy)} eV, left ratio: {left_ratio:.2f}%, right ratio: {right_ratio:.2f}%")
```

```python
def analyze_spatial_distribution(time_indexes, times, time_unit, atom_displacements, atom_displacement_magnitudes, length_unit,
                                 pristine_positions, output_prefix):
    """
    分析波包在实空间中的分布
    """
    print("\nAnalyzing wavepacket spatial distribution...")
    
    for time_idx in time_indexes:
        if time_idx >= len(times):
            continue
        disp_mags_t = atom_displacement_magnitudes[time_idx]  # shape: (natoms,)
        disps_t = atom_displacements[time_idx]  # shape: (natoms,)
        
        # 绘制位移大小的二维空间分布 (x, y, color=disp_mag)
        pristine_x = pristine_positions[:, 0]  # shape: (natoms,)
        pristine_y = pristine_positions[:, 1]  # shape: (natoms,)
        
        plot_2D_spatial_distribution(pristine_x, pristine_y, disp_mags_t*100000,
                                     rf'$x$ position ({length_unit})', 
                                     rf'$y$ position ({length_unit})',
                                     rf'Displacement ($10^{{-5}}$ {length_unit})',
                                     rf'{int(times[time_idx])} {time_unit}',
                                     output_prefix)
        
        # 绘制位移矢量的一维空间分布 (x, disps)
        #plot_1D_spatial_distribution(pristine_x, disps_t[:, 2],
        #                             rf'$x$ position ({length_unit})',
        #                             rf'Displacement magnitude ({length_unit})',
        #                             rf'{int(times[time_idx])} {time_unit}',
        #                             output_prefix)
        
    print(f"  - Spatial distribution plots saved at time ({time_unit}): {times[time_indexes]}")
```

```python
def analyze_wavepacket_propagation(time, time_unit, wavepacket_centers, length_unit, output_prefix):
    """
    分析波包在实空间中的传播
    """
    print("\nAnalyzing wavepacket propagation...")
        
    # 计算波包中心相对于初始中心的移动距离
    initial_center = wavepacket_centers[0]
    if np.any(np.isnan(initial_center)):  # 首先需要排除初始中心为 NaN
        print("  - Initial wavepacket center is NaN, cannot calculate propagation")
        return
    center_distances_from_initial = np.linalg.norm(wavepacket_centers - initial_center, axis=1)  # shape: (nframes,)
    
    
    # 寻找有效数据
    nan_indices = np.where(np.isnan(center_distances_from_initial))[0]
    if len(nan_indices) > 0:
        first_nan_index = nan_indices[0]
        valid_indices = slice(None, first_nan_index)  # 从开始到第一个 NaN (不包含)
    else:
        valid_indices = slice(None)  # 如果没有NaN, 则使用所有索引
    
    valid_time = time[valid_indices]
    valid_center_distances = center_distances_from_initial[valid_indices]
    
        
    # 尝试线性拟合来计算平均速度
    if len(valid_time) >= 2:  # 至少两个有效数据
        try:
            coeffs = np.polyfit(valid_time, valid_center_distances, 1)
            ave_speed = coeffs[0]
            print(rf"  - Average propagation speed (from center distance, with {len(valid_center_distances)} frames): {ave_speed} {length_unit}/{time_unit}")
        except np.linalg.LinAlgError:
            print("  - Could not fit a line to center distance vs time")
    else:
        print("  - No enough valid frames for propagation speed calculation")
    
    plot_time_evolution(valid_time, valid_center_distances, 
                        rf'Time ({time_unit})', rf'Wavepacket center distance ({length_unit})', r'W→N',
                        output_prefix)
```

```python
def analyze_wavepacket_decay(time, time_unit, quantity, quantity_name, quantity_unit, output_prefix):
    """
    分析某个物理量随时间的衰减
    """
    print("\nAnalyzing wavepacket decay...")
    
    print(f"  - Analyzing decay of {quantity_name}...")
    initial_value = quantity[0]
    if initial_value != 0:
        normalized_quantity = quantity / initial_value
        
        # 尝试简单的指数衰减拟合
        mask = normalized_quantity > 0  #只对正值拟合
        if np.any(mask):
            try:
                log_vals = np.log(normalized_quantity[mask])
                coeffs = np.polyfit(time[mask], log_vals, 1)
                decay_rate = -coeffs[0]
                print(f"  - Fitted decay rate (assuming exp(-rate*t)): {decay_rate} (1/{time_unit})")
            except (np.linalg.LinAlgError, ValueError):
                print("  - Could not fit exponential decay")
        else:
            print("  - No positive values for exponential fit")
        
        # 绘制衰减曲线
        plot_time_evolution(time, normalized_quantity,
                            rf'Time ({time_unit})', rf'{quantity_name} (Norm.)', r'W→N',
                            output_prefix)

    else:
        print(f"  - Cannot analyze decay of {quantity_name}, initial value is zero")
```

```python
def analyze_wavepacket_from_trajectory(pristine_structure_without_wavepacket_file='model_without_wavepacket.xyz',
                                       structure_0_file = 'model_with_wavepacket.xyz',
                                       trajectory_file='dump.xyz', dump_interval=1,
                                       wavepacket_center_0=None, output_prefix='wavepacket_analysis'):
    """
    分析 GPUMD dump_exyz 输出的轨迹文件, 提取波包信息
    
    Args:
        pristine_structure_without_wavepacket_file (str): 不包含波包的原始结构文件 (default: 'model_without_wavepacket.xyz')
        structure_0_file (str): t=0时刻的结构文件 (default: 'model_with_wavepacket.xyz')
        trajectory_file (str): extxyz格式的轨迹文件 (default: 'dump.xyz')
        dump_interval (float): MD模拟的时间步, 单位: fs (default; 1) 
        wavepacket_center_0 (list or None): t=0时刻的波包中心位置 [x, y, z], 单位: A; 如果为 None, 则使用初始结构的几何中心 (default: None)
        output_prefix (str): 输出文件的前缀 (default: 'wavepacket_analysis') 
    """
    print("\nAnalyzing phonon wavepacket from molecular dynamics simulations...")
    
    # 读取初始结构
    print(f"\nReading pristine structure without wavepacket from {pristine_structure_without_wavepacket_file} file...") 
    
    pristine_atoms = read(pristine_structure_without_wavepacket_file, format='extxyz')  # 默认结构文件为 extxyz 格式
    natoms = len(pristine_atoms)
    print(f"  - Number of atoms: {natoms}")
    
    masses = pristine_atoms.get_masses()
    pristine_positions = pristine_atoms.get_positions()
    
    if 'energy' in pristine_atoms.info:
        pristine_potential_energy = pristine_atoms.info.get('energy')  # unit: eV
    else:
        print(f"  - Warning: potential energy not found in pristine structure, set to NaN")
        pristine_potential_energy = np.nan
    print(f"  - Pristine potential energy (eV): {pristine_potential_energy}")
    
    if 'energy_atom' in pristine_atoms.arrays:
        pristine_atom_potential_energies = pristine_atoms.arrays['energy_atom']  # unit: eV, shape: (natoms,)
    else:
        print(f"  - Warning: atom potential energy not found in in pristine structure, set to NaN")
        pristine_atom_potential_energies = np.full(natoms, np.nan)

    
    # 读取轨迹文件
    print(f"\nReading trajectory from {structure_0_file} + {trajectory_file} file...")
    
    ## 存储需要的物理量
    times = []
    mean_squared_displacements_per_frame = []
    atom_displacements_per_frame = []
    atom_displacement_magnitudes_per_frame = []  # 存储每个原子在每个时间步的位移大小
    mean_squared_velocities_per_frame = []
    atom_velocity_magnitudes_per_frame = []  # 存储每个原子在每个时间不的速度大小
    
    kinetic_energies_per_frame = []
    potential_energies_per_frame = []
    atom_potential_energies_per_frame = []
    
    wavepacket_centers_per_frame = []
    
    ## 创建一个生成器, 将初始结构和轨迹帧合并
    def combined_frames(atoms_0, trajectory_file):
        yield atoms_0  # 首先 yield 初始结构
        for frame in iread(trajectory_file, format='extxyz'):  # 然后 yield 轨迹文件中的每一帧
            yield frame
    
    ## 使用 ase.io.iread 逐帧读取轨迹, 节省内存
    frame_count = 0
    atoms_0 = read(structure_0_file, format='extxyz')
    for frame in combined_frames(atoms_0, trajectory_file):
        frame_count += 1
        
        # 读取当前帧的物理时间
        if 'Time' in frame.info:  # Atoms.info 是一个字典
            time = frame.info.get('Time')  # unit: fs
        else:
            print(f"  - Warning: time not found in frame {frame_count}, calculated by dump_interval")
            time = (frame_count - 1) * dump_interval  # 如果帧中没有时间信息, 需要时间步长 dump_interval 来计算当前帧的时间
        times.append(time)
        
        # 计算当前帧的位移 (相对于原始结构)
        pos_t = frame.get_positions()  # unit: A
        disp_t = pos_t - pristine_positions  # unit: A
        atom_displacements_per_frame.append(disp_t)
        
        atom_disp_mags_t = np.linalg.norm(disp_t, axis=1)  # shape: (natoms,)
        atom_displacement_magnitudes_per_frame.append(atom_disp_mags_t)
        
        mean_squared_displacement_t = np.mean(np.sum(disp_t**2, axis=1))
        mean_squared_displacements_per_frame.append(mean_squared_displacement_t)
        
        # 读取当前帧的速度 (如果轨迹文件包含)
        if 'vel' in frame.arrays:
            vel_t = frame.arrays['vel']   # unit: A/fs
        else:
            print(f"  - Warning: velocity not found in frame {frame_count}, set to zero")
            vel_t = np.zeros_like(pos_t)  # 如果没有速度, 则为零
            
        atom_vel_mags_t = np.linalg.norm(vel_t, axis=1)  # shape: (natoms,)
        atom_velocity_magnitudes_per_frame.append(atom_vel_mags_t)
        
        mean_squared_velocity_t = np.mean(np.sum(vel_t**2, axis=1))
        mean_squared_velocities_per_frame.append(mean_squared_velocity_t)

        # 计算当前帧的动能
        kinetic_energy_t_amu_A2_fs2 = 0.5 * np.sum(masses * np.sum(vel_t**2, axis=1))  # unit: amu * (A/fs)^2
        # 或使用 ase.Atoms.get_kinetic_energy() 方法获取动能
        #frame.set_masses(masses)
        #frame.set_velocities(vel_t)
        #kinetic_energy_t_amu_A2_fs2 = frame.get_kinetic_energy()
        kinetic_energy_t_eV = kinetic_energy_t_amu_A2_fs2 * 103.6427  # unit: eV
        kinetic_energies_per_frame.append(kinetic_energy_t_eV)

        # 读取势能 (如果轨迹文件包含)
        if 'energy' in frame.info:
            potential_energy_t = frame.info.get('energy')  # unit: eV
        else:
            print(f"  - Warning: potential energy not found in frame {frame_count}, set to NaN")
            potential_energy_t = np.nan
        potential_energies_per_frame.append(potential_energy_t)
        
        # 读取原子势能 (如果轨迹文件包含)
        if 'energy_atom' in frame.arrays:
            atom_potential_energies_t = frame.arrays['energy_atom']  # unit: eV, shape: (natoms,)
        else:
            print(f"  - Warning: atom potential energy not found in frame {frame_count}, set to NaN")
            atom_potential_energies_t = np.full(natoms, np.nan)
        atom_potential_energies_per_frame.append(atom_potential_energies_t)
            
        # 计算当前帧的波包中心
        if frame_count == 1:
            wavepacket_center_t = np.array(wavepacket_center_0)  # t=0 使用初始中心
        else:
            x_center_fit, y_center_fit, z_center_fit, A_fit, sigma_fit, fit_success = fit_gaussian_envelope_3d(pristine_positions, atom_disp_mags_t, tolerance=0.01)
            if fit_success:
                wavepacket_center_t = np.array([x_center_fit, y_center_fit, z_center_fit])
                print(f"  - Fitted 3D wavepacket center at frame {frame_count}: ({x_center_fit:.3f}, {y_center_fit:.3f}, {z_center_fit:.3f})")
            else:
                print(f"  - Failed to get valid 3D wavepacket center at frame {frame_count} (fit failed or center out of range), set to NaN")
                wavepacket_center_t = np.array([np.nan, np.nan, np.nan])     
        wavepacket_centers_per_frame.append(wavepacket_center_t)

    print(f"  - Read {frame_count} frames from trajectory (including t=0)")
    print(f"  - Total {max(times)} fs in trajectory")
    
    times = np.array(times)
    atom_displacements_per_frame = np.array(atom_displacements_per_frame)  # shape: (nframes, natoms, 3)
    atom_displacement_magnitudes_per_frame = np.array(atom_displacement_magnitudes_per_frame)  # shape: (nframes, natoms)
    mean_squared_displacements_per_frame = np.array(mean_squared_displacements_per_frame)

    atom_velocity_magnitudes_per_frame = np.array(atom_velocity_magnitudes_per_frame)  # shape: (nframes, natoms)
    mean_squared_velocities_per_frame = np.array(mean_squared_velocities_per_frame)
    
    kinetic_energies_per_frame = np.array(kinetic_energies_per_frame)
    potential_energies_per_frame = np.array(potential_energies_per_frame)
    atom_potential_energies_per_frame = np.array(atom_potential_energies_per_frame)
    wavepacket_centers_per_frame = np.array(wavepacket_centers_per_frame)

    
    # 分析波包空间分布
    time_indexes = range(0, 21, 10)
    analyze_spatial_distribution(time_indexes, times, 'fs', 
                                 atom_displacements_per_frame, atom_displacement_magnitudes_per_frame, r'$\mathrm{\AA}$',
                                 pristine_positions, output_prefix)
    
    
    # 计算能量透射
    calculate_energy_transmission(time_indexes, times, 'fs', 
                                          atom_velocity_magnitudes_per_frame, atom_potential_energies_per_frame,
                                          pristine_atom_potential_energies, masses,
                                          pristine_positions, 'x', interface_position=None)
    
    
    # 分析波包传播
    analyze_wavepacket_propagation(times, 'fs', wavepacket_centers_per_frame, r'$\mathrm{\AA}$', output_prefix) 
    
    
    # 分析波包衰减
    ## 势能的衰减
    if not np.all(np.isnan(kinetic_energies_per_frame)):
        analyze_wavepacket_decay(times, 'fs', kinetic_energies_per_frame, 'Kinetic energy', 'eV', output_prefix)
    else:
        print("  - No total energy data found for decay analysis")
    
    ## 最大原子位移的衰减
    analyze_wavepacket_decay(times, 'fs', np.max(atom_displacement_magnitudes_per_frame, axis=1), 'Maximal displacement', r'$\mathrm{\AA}$', output_prefix)
    
    print("\nWavepacket analysis completed")
```
